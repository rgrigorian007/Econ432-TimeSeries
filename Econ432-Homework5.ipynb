# Part I: Review Questions

## Problem 1

One popular concept on stock market price is called "martingale pricing", which is relevant to "rational expectations" or "efficient market hypothesis" in economic theories. A rough statement on this is that, a (log of) stock market price $p_t$ (e.g., S&P 500 index) is supposed to reflect all the relevant information available up to time $t$, i.e.,
$$\mathbb{E}[p_t | \mathcal{F}_{t-1}] = p_{t-1}$$
therefore $p_t = \log P_t$ is a martingale.

### Part (a)

Under martingale pricing, show that continuously compounded (cc) return for stock market priace is a martingale difference sequence (mds).

#### Proof

By definition, $\{d_t \}_{t=1}^T$ is a martingale difference sequence if $\mathbb{E} [ d_t | \mathcal{F}_{t-1}] = 0$ where $d_t \equiv Y_t - Y_{t-1}$. In our case, cc returns are defined as
$$r_t = \ln \left( \frac{P_t}{P_{t-1}} \right) = p_t - p_{t-1}$$
Therefore, applying the definition of an mds
$$\mathbb{E}[r_t | \mathcal{F}_{t-1}] = \mathbb{E}[p_t - p_{t-1} | \mathcal{F}_{t-1}] = \mathbb{E} [p_t | \mathcal{F}_{t-1}] - \mathbb{E}[p_{t-1} | \mathcal{F}_{t-1}]$$
By definition of a martingale and by properties of $\sigma$ fields
$$\mathbb{E} [p_t | \mathcal{F}_{t-1}] - \mathbb{E}[p_{t-1} | \mathcal{F}_{t-1}] = p_{t-1} - p_{t-1} = 0$$
Therefore, cc returns, $r_t$ are a MDS
\begin{align*}
\tag*{$\blacksquare$}
\end{align*}

### Part (b)

In what follows, we assume cc returns $r_t$ is a covariance stationary process. Prove the following statements:

1. If $r_t \sim iid(0, \sigma^2)$ or (independent white noise), then $r_t \sim mds(0, \sigma^2)$.

2. If $r_t \sim mds(0,\sigma^2)$, then $r_t \sim WN (0, \sigma^2)$ (or weak white noise).

#### Proof

##### Statement 1

$r_t \sim iid(0, \sigma^2)$ implies that $r_t$ is independent and identically distributed. If $r_t$ is independent of $r_{t-j} \ \forall j$, then $\mathbb{E}[r_t | \mathcal{F}_{t-1}] = 0$. Past information has no affect on a future event. By definition, this is a martingale difference sequence.
\begin{align*}
\tag*{$\blacksquare$}
\end{align*}

##### Statement 2

For $r_t$ to be $WN (0, \sigma^2)$, we need to show:
$$\mathbb{E}[r_t] = 0, \quad \operatorname{Var}(r_t) = \sigma^2 , \quad \operatorname{Cov}(r_t, r_{t-j}) = 0$$
If $r_t \sim mds(0, \sigma^2)$, then $\mathbb{E}[r_t | \mathcal{F}_{t-1}] = 0$ and $\operatorname{Var}(r_t) =\mathbb{r_t^2} = \sigma^2$ for all $t$. Therefore, show the first requirement by Law of Iterated Expectations:
$$\mathbb{E}\Big[ \mathbb{E}[r_t | \mathcal{F}_{t-1}]\Big] = \mathbb{E}[0] \implies \mathbb{E}[r_t] = 0$$
The second requirement is trivially satisfied by definition of our MDS. And the last condition is satisified by
\begin{align*}
\operatorname{Cov}(r_t, r_{t-j}) &= \mathbb{E}[r_t \cdot r_{t-j}] - \mathbb{E}[r_t]\mathbb{E}[r_{t-j}] \\
&= \mathbb{E}[r_t \cdot r_{t-j}] \\
&= \mathbb{E} \Big[\mathbb{E}[r_t \cdot r_{t-j} | \mathcal{F}_{t-1}] \Big] \quad \quad \textrm{(Law of Iterated Exp.)} \\
&= \mathbb{E}\Big[ r_{t-j} \mathbb{E}[r_t | \mathcal{F}_{t-1}]\Big] \\
&= \mathbb{E}[r_{t-j} \cdot 0] = 0
\end{align*}
Hence, all three conditions are satisfied.
\begin{align*}
\tag*{$\blacksquare$}
\end{align*}

### Part (c)

Prove the following statements:

1. If $\{r_t\}$ is iid, then it is strictly stationary.

2. If $\{r_t\}$ is strictly stationary and $\mathbb{E}[r_1^2] < \infty$, then it is covariance stationary.

#### Proof

##### Statement 1

If $\{r_t\}$ is iid, then
$$f(r_{t_1}, r_{t_2}, \dots , r_{t_k}) = f_1 (r_{t_1}) \cdot f_2 (r_{t_2}) \cdot \quad \cdots \quad \cdot f_t (r_{t_k})$$
by property of independence. And then by property of identical distribution
$$f(r_{t_1}, r_{t_2}, \dots , r_{t_k}) = f(r_{1}) \cdot f(r_{2}) \cdot \quad \cdots \quad \cdot f(r_{k})$$
Therefore,
$$f(r_{t_1 + \eta}, r_{t_2 + \eta}, \dots , r_{t_k + \eta}) = f(r_{1}) \cdot f(r_{2}) \cdot \quad \cdots \quad \cdot f(r_{k})$$
for any $\eta \in \mathbb{N}$. Hence $\{r_t\}$ is strictly stationary.
\begin{align*}
\tag*{$\blacksquare$}
\end{align*}

##### Statement 2

To prove covariance stationarity, we need to show that:

1. $\mathbb{E}[| r_t^2 |] < \infty \quad \forall t \in T$
2. $\mathbb{E}[r_t] = \mu$ and $\operatorname{Var}(r_t) = \sigma^2 \quad \forall t \in T$
3. $\operatorname{Cov}(r_t, r_{t-j}) = \gamma_j$

By assumption $\mathbb{E}[r_1^2] < \infty$, and since strict stationarity implies identical distribution across all time $t$, then condition 1 is trivially satisified. Condition 2 is satisfied by properties of strict stationarity which imply time invariance for *all* higher moments if they exist. 

To show condition 3, want to prove that the autocovariance of our time series is time invariant. Hence, we will show that
$$\operatorname{Cov}(r_t, r_{t-j}) = \operatorname{Cov}(r_{t+s}, r_{t+s-j}) \quad \forall \  t,j,s\in T \subseteq \mathbb{N}$$
By expectation expansion of covariance,
\begin{align*}
\operatorname{Cov}(r_t, r_{t-j}) &= \mathbb{E}\Big[(r_t - \mathbb{E}[r_t])(r_{t-j} - \mathbb{E}[r_{t-j}])\Big] = \mathbb{E}[r_t \cdot r_{t-j}] - \mu^2 \\
\operatorname{Cov}(r_{t+s}, r_{t+s-j}) &= \mathbb{E}\Big[(r_{t+s} - \mathbb{E}[r_{t+s}])(r_{t+s-j} - \mathbb{E}[r_{t+s-j}])\Big] = \mathbb{E}[r_{t+s} \cdot r_{t+s-j}] - \mu^2
\end{align*}
where by strict stationarity, the first moment is time invariant. These expressions then implies that
$$\operatorname{Cov}(r_t, r_{t-j}) = \operatorname{Cov}(r_{t+s}, r_{t+s-j}) \iff \mathbb{E}[r_t \cdot r_{t-j}] = \mathbb{E}[r_{t+s} \cdot r_{t+s-j}]$$
By the integral expression of expectation, we show
\begin{align*}
\mathbb{E}[r_t \cdot r_{t-j}] &= \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} r_{t}r_{t-j}f(r_{t}, r_{t-j})dr_{t} dr_{t-j} \\
\mathbb{E}[r_{t+s} \cdot r_{t+s-j}] &= \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} r_{t+s}r_{t+s-j}f(r_{t+s}, r_{t+s-j})dr_{t+s} dr_{t+s-j}
\end{align*}
Then by the definition of strict stationarity, $f(r_{t+s}, r_{t+s-j}) = f(r_{t}, r_{t-j})$ which then implies the equivalence
\begin{align*}
\mathbb{E}[r_t \cdot r_{t-j}] &=\int_{-\infty}^{\infty}\int_{-\infty}^{\infty} r_{t}r_{t-j}f(r_{t}, r_{t-j})dr_{t} dr_{t-j} \\
&= \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} r_{t+s}r_{t+s-j}f(r_{t}, r_{t-j})dr_{t+s} dr_{t+s-j} \\
&= \mathbb{E}[r_{t+s} \cdot r_{t+s-j}]
\end{align*}
where the uniqueness of the integral is determined by the density function. The other variables are simply the variables of integration and can theoretically by anything (e.g., the principle of U-substitution). Therefore, since these expectations are equivalent, then the covariances are equivalent as well. Therefore, covariances are equivalent across time -- time invariant $\operatorname{Cov}(r_t, r_{t-j}) = \gamma_j$.
\begin{align*}
\tag*{$\blacksquare$}
\end{align*}

## Problem 2

Suppose that $\varepsilon_t \sim iid \ N(0,1)$. Define $Y_{1,t} \equiv \varepsilon_t \varepsilon_{t-1}$ and $Y_{2,t} \equiv \varepsilon_t + \varepsilon_{t-1}\varepsilon_{t-2}$. Show the following.

### Part (a)

Prove $Y_{1,t}$ is an mds but not iid.

#### Proof

A stochastic process is a martingale difference sequence if the conditional expectation with respect the $\sigma$-field $\mathcal{F}_{t-1}$ is equal to 0. Hence,
$$\mathbb{E}[Y_{1,t} | \mathcal{F}_{t-1}] = \mathbb{E}[\varepsilon_t \varepsilon_{t-1} | \mathcal{F}_{t-1}] = \varepsilon_{t-1} \mathbb{E}[\varepsilon_t] = \varepsilon \cdot 0 = 0$$
and we confirm that $Y_{1,t}$ is a mds. We show that it is not iid by showing that $Y_{1,t}^2$ and $Y_{1,t-1}^2$ are correlated.
\begin{align*}
\operatorname{Cov}(Y_{1,t}^2 , Y_{1,t-1}^2) &= \mathbb{E}\Big[ Y_{1,t}^2 \cdot Y_{1,t-1}^2 \Big] -\mathbb{E}[Y_{1,t}^2] \mathbb{E}[Y_{1,t-1}^2] \\
&=\mathbb{E}\Big[\varepsilon_t^2 \varepsilon_{t-1}^2 \varepsilon_{t-1}^2  \varepsilon_{t-2}^2 \Big] - \mathbb{E}\Big[ \varepsilon_t^2 \varepsilon_{t-1}^2 \Big]\mathbb{E}\Big[ \varepsilon_{t-1}^2 \varepsilon_{t-2}^2 \Big] \\
&=\mathbb{E}\Big[\varepsilon_t^2 \varepsilon_{t-1}^4 \varepsilon_{t-2}^2 \Big] -\mathbb{E}\Big[\varepsilon_{t}^2 \Big]\mathbb{E}\Big[\varepsilon_{t-1}^2 \Big]\mathbb{E}\Big[\varepsilon_{t-1}^2 \Big]\mathbb{E}\Big[\varepsilon_{t-2}^2 \Big] \\
&= \mathbb{E}\Big[ \varepsilon_t^2 \Big]\mathbb{E}\Big[ \varepsilon_{t-1}^4 \Big]\mathbb{E}\Big[ \varepsilon_{t-2}^2 \Big] - \mathbb{E}\Big[\varepsilon_{t}^2 \Big]\mathbb{E}\Big[\varepsilon_{t-1}^2 \Big]\mathbb{E}\Big[\varepsilon_{t-1}^2 \Big]\mathbb{E}\Big[\varepsilon_{t-2}^2 \Big] \\
&= 1 \cdot 3 \cdot 1 - 1 \cdot 1 \cdot 1 \cdot 1 \\
&= 3 - 1 = 2 \neq 0
\end{align*}
where we know the fourth moment of a standard normal variable is the kurtosis which is equal to 3. Hence, we show that $Y_{1,t}$ is mds but not iid
\begin{align*}
\tag*{$\blacksquare$}
\end{align*}

### Part (b)

Prove $Y_{2,t}$ is a WN but not a mds.

#### Proof

$Y_{2,t}$ is a White Noise process if $\mathbb{E}[Y_{2,t}] = 0$, $\operatorname{Var}(Y_{2,t}) = \sigma^2$, and $\operatorname{Cov}(Y_{2,t}, Y_{2,s}) = 0$ for all $t \neq s$. We show the first condition:
$$\mathbb{E}[Y_{2,t}] = \mathbb{E}[\varepsilon_t + \varepsilon_{t-1}\varepsilon_{t-2}] = 0 + \mathbb{E}[\varepsilon_{t-1}] \mathbb{E}[\varepsilon_{t-2}] = 0$$
By the fact that $\varepsilon_t$ is iid for all t. We can show the second condition (constant variance):
\begin{align*}
\operatorname{Var}(Y_{2,t}) &= \mathbb{E} \Big[ (Y_{2,t})^2 \Big] - \overbrace{\mathbb{E}[Y_{2,t}]^2}^{=0} \\
&= \mathbb{E}\Big[ (\varepsilon_{t} + \varepsilon_{t-1}\varepsilon_{t-2})^2 \Big] \\
&= \mathbb{E}[\varepsilon_{t}^2] + 2\mathbb{E}[\varepsilon_{t} \varepsilon_{t-1} \varepsilon_{t-2}] + \mathbb{E}[\varepsilon_{t-1}^2]\mathbb{E}[\varepsilon_{t-2}^2] \\
&= \operatorname{Var}(\varepsilon_{t}) + \mathbb{E}[\varepsilon_{t}]\mathbb{E}[\varepsilon_{t-1}\varepsilon_{t-2}] + \operatorname{Var}(\varepsilon_{t-1})\operatorname{Var}(\varepsilon_{t-2}) \\
&= 1 + 0 + 1 = 2
\end{align*}
Hence, the variance is a constant $\sigma^2 = 2$ for all $t$. And lastly, to show that $Y_{2,t}$ is a white noise process, we must show that the autocovariance is 0 across time.
\begin{align*}
\operatorname{Cov}(Y_{2,t}, Y_{2,s}) &= \mathbb{E}\Big[ Y_{2,t} \cdot Y_{2,s} \Big] - \mathbb{E}[Y_{2,t}]\mathbb{E}[Y_{2,s}]\\
&= \mathbb{E} \Big[ (\varepsilon_{t} + \varepsilon_{t-1}\varepsilon_{t-2} )(\varepsilon_{s} + \varepsilon_{s-1}\varepsilon_{s-2})\Big] - 0 \\
&= \mathbb{E}\Big[\varepsilon_{t}\varepsilon_{s} + \varepsilon_{t}\varepsilon_{s-1}\varepsilon_{s-2} \varepsilon_{s}\varepsilon_{t-1}\varepsilon_{t-2} + \varepsilon_{t-1}\varepsilon_{t-2}\varepsilon_{s-1}\varepsilon_{s-2}\Big] \\
&= 0 \quad \quad (\textrm{By iid})
\end{align*}
Therefore, the autocovariance is 0 across time. We have shown that $Y_{2,t}$ is a white noise process. To show that it is **not** an mds we simply show that the conditional expectation is not equal to zero.
$$\mathbb{E}[Y_{2,t} | \mathcal{F}_{t-1}] = \mathbb{E}[\varepsilon_t + \varepsilon_{t-1}\varepsilon_{t-2} | \mathcal{F}_{t-1}] = 0 + \mathbb{E}[\varepsilon_{t-1}\varepsilon_{t-2} | \mathcal{F}_{t-1}] = \varepsilon_{t-1}\varepsilon_{t-2} \neq 0$$
Hence, the process is not a mds.
\begin{align*}
\tag*{$\blacksquare$}
\end{align*}

## Problem 3

Suppose that $\{Y_t\}$ is a covariance-stationary process generated from the AR(1) model
$$Y_t - \mu = \phi(Y_{t-1} - \mu) + \varepsilon_t$$
where $|\phi| < 1$ and $\varepsilon_t \sim iid(0, \sigma^2_\varepsilon)$.

### Part (a)

Show that $\mathbb{E}[Y_t] = \mu$.

#### Proof

We begin by taking the expectation of both sides
\begin{align*}
\mathbb{E}[Y_t - \mu] &= \mathbb{E}\Big[ \phi (Y_{t-1} - \mu) + \varepsilon_t \Big] \\
\mathbb{E}[Y_t] - \mu &= \phi \mathbb{E}[Y_{t-1}] - \phi \mu + 0 \\
\mathbb{E}[Y_t] &= \phi \mathbb{E}[Y_{t-1}] - \phi \mu + \mu
\end{align*}
We can then rearrange to get
\begin{align*}
\mathbb{E}[Y_t] - \phi \mathbb{E}[Y_{t-1}] &= - \phi \mu + \mu \\
\mathbb{E}[Y_t] (1 - \phi) &= \mu (1 - \phi)
\end{align*}
Which then implies
\begin{align*}
\mathbb{E}[Y_t] &= \frac{\mu (1- \phi)}{1-\phi} = \mu \tag*{$\blacksquare$}
\end{align*}

### Part (b)

Show that $\operatorname{Cov}(Y_{t-j} - \mu, \varepsilon_t) = 0$ for any $j \ge 1$.

#### Proof

We utilize the expectation represetnation of Covariance to show
\begin{align*}
\operatorname{Cov}(Y_{t-j} - \mu, \varepsilon_t) &= \mathbb{E}\Big[ (Y_{t-j} - \mu)\varepsilon_t \Big] - \mathbb{E}[Y_{t-j} - \mu] \overbrace{\mathbb{E}[\varepsilon_t]}^{=0} \\
&= \mathbb{E}[Y_{t-j}\varepsilon_t - \mu \varepsilon_t] = \mathbb{E}[Y_{t-j}\varepsilon_t] - \mu \mathbb{E}[\varepsilon_t] \\
&= \mathbb{E}[Y_{t-j}\varepsilon_t] = \mathbb{E}[Y_{t-j}]\mathbb{E}[\varepsilon_t] \\
&= 0
\end{align*}
We can separate that last expectation since the error term is independent of the time series.
\begin{align*}
\tag*{$\blacksquare$}
\end{align*}

### Part (c)

Show that $\operatorname{Var}(Y_t) = \sigma^2_\varepsilon (1 - \sigma^2)^{-1}$.

#### Proof

We can take the variance of both sides of our AR(1) model
\begin{align*}
\operatorname{Var}(Y_t - \mu) &= \operatorname{Var}(\phi(Y_{t-1} - \mu) + \varepsilon_t) \\
\end{align*}
Variance is applied linearly across addition/subtraction, and the variance of a constant $\mu$ is 0. Hence
$$\operatorname{Var}(Y_t) = \operatorname{Var}(\phi Y_{t-1} - \phi \mu + \varepsilon_t)$$
where by properties of covariance-stationary processes, $\operatorname{Var}(Y_t) = \operatorname{Var}(Y_{t-1})$. Therefore,
$$\operatorname{Var}(Y_t) = \phi^2 \operatorname{Var}(Y_t) + \operatorname{Var}(\varepsilon_t)$$
Subtracting the variance of $Y_t$ term from the left hand side and dividing by $(1-\phi)^2$ shows
$$\operatorname{Var}(Y_t) = \sigma^2_\varepsilon (1 - \sigma^2)^{-1}$$

\begin{align*}
\tag*{$\blacksquare$}
\end{align*}

### Part (d)

Recall that $\gamma_j = \operatorname{Cov}(Y_t , Y_{t-j})$. Show that $\gamma_j = \phi \gamma_{j-1}$.

#### Proof

We have the base (mean adjusted) AR(1) model:
$$Y_t - \mu = \phi(Y_{t-1} - \mu) + \varepsilon_t$$
By simple properties of multiplication, we can multiply both sides of the expression by the same value. In this case, we multiply both sides by $(Y_{t-j} - \mu) to get
$$(Y_{t-j} - \mu) (Y_t - \mu) = (Y_{t-j} - \mu)(\phi(Y_{t-1} - \mu) + \varepsilon_t)$$
We can then take the expectations of both sides to prove our claim
\begin{align*}
\mathbb{E}\Big[ (Y_{t-j} - \mu) (Y_t - \mu) \Big] &= \mathbb{E}\Big[ (Y_{t-j} - \mu)(\phi(Y_{t-1} - \mu) + \varepsilon_t) \Big]
\end{align*}
where the left hand side (by definition) is equal to the relevant covariance term
\begin{align*}
\operatorname{Cov}(Y_t , Y_{t-j}) &= \mathbb{E}\Big[ (Y_{t-j} - \mu)(\phi(Y_{t-1} - \mu) + \varepsilon_t) \Big] \\
&= \mathbb{E}\Big[ (Y_{t-j} - \mu)(\phi(Y_{t-1} - \mu))+ (Y_{t-j} - \mu)\varepsilon_t \Big] \\
&= \mathbb{E}\Big[ (Y_{t-j} - \mu)(\phi(Y_{t-1} - \mu))\Big] + \mathbb{E}[(Y_{t-j} - \mu)\varepsilon_t] \\
&= \phi \operatorname{Cov}(Y_{t-1} , Y_{t-j}) + 0
\end{align*}
And by definitions $\gamma_j \equiv \operatorname{Cov}(Y_t , Y_{t-j})$ and $\gamma_{j-1} \equiv \operatorname{Cov}(Y_{t-1} , Y_{t-j})$ we show
\begin{align*}
\gamma_j = \phi \gamma_{j-1} \tag*{$\blacksquare$}
\end{align*}

### Part (e)

Using part (c) and part (d), show that $\gamma_j = \phi^j \sigma^2_\varepsilon (1 - \phi^2)^{-1}$.

#### Proof

By definition of covariance, we know that $\operatorname{Cov}(Y_t , Y_{t}) = \operatorname{Var}(Y_t) = \gamma_0$. Hence, we use the result from part (c) and (d) to show
\begin{align*}
\gamma_j &= \phi \gamma_{j-1} \\
&= \phi (\phi \gamma_{j-2}) \\
&\quad \vdots \\
&= \phi^j \gamma_{j- j} = \phi^j \gamma_0 \\
&= \phi^j \operatorname{Var}(Y_t) \\
&= \phi^j \sigma^2_\varepsilon (1 - \phi^2)^{-1} \tag*{$\blacksquare$}
\end{align*}

### Part (f)

Recall that $\rho_j = \gamma_j / \gamma_0$. Show that $\rho_j = \phi^j$.

#### Proof

By definition and using our result from part (e)
$$\rho_j = \frac{\gamma_j}{\gamma_0} = \frac{\phi^j \gamma_0}{\gamma_0} = \phi^j$$

\begin{align*}
\tag*{$\blacksquare$}
\end{align*}

## Problem 4

Consider the following AR(2) model:
$$Y_t = \mu + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \varepsilon_t$$
where $\varepsilon \sim iid \ N(0, \sigma^2_\varepsilon)$.

### Part (a)

Show that the above expression can be written as
$$Y_t = \mu + \rho Y_{t-1} + \alpha_1 \Delta Y_{t-1} + \varepsilon_t$$
where $\Delta Y_{t-1} = Y_{t-1} - Y_{t-2}$.

#### Proof

If $\Delta Y_{t-1} = Y_{t-1} - Y_{t-2}$, then we can write $Y_{t-2} = Y_{t-1} - \Delta Y_{t-1}$. We can then substitute this expression into our AR(2) model:
\begin{align*}
Y_t &= \mu + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \varepsilon_t \\
&= \mu + \phi_1 Y_{t-1} + \phi_2 (Y_{t-1} - \Delta Y_{t-1}) + \varepsilon_t \\
&= \mu + (\phi_1 + \phi_2)Y_{t-1} - \phi_2 \Delta Y_{t-1} + \varepsilon_t \\
Y_t &= \mu + \rho Y_{t-1} + \alpha_1 \Delta Y_{t-1} + \varepsilon_t
\end{align*}
where $\rho \equiv \phi_1 + \phi_2$ and $\alpha_1 \equiv -\phi_2$.
\begin{align*}
\tag*{$\blacksquare$}
\end{align*}

### Part (b)

Using the daily log-prices of Apple in 2021 for $Y_t$, we run a linear regression on the expression from part (a) and find the following:
$$Y_t = \underset{(0.0497)}{0.006} + \underset{(0.0101)}{0.9991} \ Y_{t-1} - \underset{(0.0640)}{0.0386} \ \Delta Y_{t-1} + \hat{\varepsilon}_t$$
where the numbers in parentheses are the standard errors reported in the LS regression. We are interested in testing the null hypothesis that the daily log-prices of Apple is a unit root process. Does the augmented Dickey-Fuller (ADF) test support this null under the significance level of 0.05?

#### Solution

We want to test the set of hypotheses
$$H_0 \ : \ \rho = 1 \quad \textrm{v.s.} \quad H_1 \ : \ \rho \neq 1$$
using the Augmented Dickey Fuller (ADF) test. With the above regression results, we compute the ADF statistic
$$\textrm{ADF} = \frac{\hat{\rho} - 1}{\operatorname{SE}(\hat{\rho})} = \frac{0.9991 - 1}{0.0101} \approx -0.0891 > -2.8604$$
where $\textrm{DF}_{0.05} = -2.8604$ is our critical value. In this case, since our test statistic is less negative than our critical value, we **fail to reject the null**. Therefore, the ADF test supports the null (unit root process) under the 5% significance level.

### Part (c)

Now suppose that $\rho = 1$. Show that the expression in part (a) can be written as
$$\Delta Y_t = \mu + \alpha_1 \Delta Y_{t-1} + \varepsilon_t$$
where $\Delta Y_t = Y_t - Y_{t-1}$. Using the same data in part (b), we obtain the following results from linear regression of the above expression:
$$\Delta Y_t = \underset{(0.0010)}{0.0014} - \underset{(0.0631)}{0.0395} \ \Delta Y_{t-1} + \hat{\varepsilon}_t$$
We want to test if the daily cc return of Apple is a unit-root process. Does the Dickey-Fuller (DF) test support this null under the significance level of 0.05?

#### Solution

With $\rho = 1$, we can rewrite our expression as
\begin{align*}
Y_t &= \mu + \rho Y_{t-1} + \alpha_1 \Delta Y_{t-1} + \varepsilon_t \\
&= \mu + Y_{t-1} + \alpha_1 \Delta Y_{t-1} + \varepsilon_t \\
Y_t - Y_{t-1} &= \mu + \alpha_1 \Delta Y_{t-1} + \varepsilon_t \\
\implies \Delta Y_t &= \mu + \alpha_1 \Delta Y_{t-1} + \varepsilon_t \tag*{$\blacksquare$}
\end{align*}
Now, we can test the set of hypotheses
$$H_0 \ : \ \alpha_1 = 1 \quad \textrm{v.s.} \quad H_1 \ : \ \alpha_1 \neq 1$$
using the Dickey Fuller (DF) test. With the above regression results, we compute the DF statistic
$$\textrm{DF} = \frac{\hat{\alpha}_1 - 1}{\operatorname{SE}(\hat{\alpha}_1)} = \frac{-0.0395 - 1}{0.0631} \approx -16.464 < -2.8604$$
where $\textrm{DF}_{0.05} = -2.8604$ is our critical value. In this case, since our test statistic is more negative than our critical value, we **reject the null**. Therefore, the DF test does not support the null (unit root process) under the 5% significance level.

```python
# Define Plotting Function
def ma1_plot(theta):
    ar = np.r_[1] # add zero-lag
    ma = np.r_[1, theta] # add zero-lag
    y = 0.05 + arma_generate_sample(ar, ma, scale = 0.1, nsample = 250)
    
    fig, ax = plt.subplots(nrows = 3, ncols = 1,figsize=(8,12))
    ax[0].plot(np.arange(1, 251, 1), y) # first subgraph
    ax[0].title.set_text(
        r'MA(1) Process with $\theta$ = {}'.format(theta)) # add the title
    
    theoretical_acf = [1, theta/(1+theta ** 2)] + [0] * 14
    ax[1].scatter(np.arange(0, 16, 1), theoretical_acf)
    ax[1].set_ylim([-1.0, 1.0])
    ax[1].title.set_text('Theoretical ACF')
    
    plot_acf(y, lags = 15, ax = ax[2]) # second subgraph
    ax[2].title.set_text('Sample ACF')
    plt.show()
```

### Part (a)

Use the Python function `statsmodels.tsa.arima_process.arma_generate_sample()` to simulate and plot 250 observations of the AR(1) with $\phi = 0$, $\phi = 0.5$, $\phi = 0.9$, $\phi = 0.99$.

#### Solution

```python
# Define Plotting Function
def ar1_plot(phi):
    ar = np.r_[1, -phi] # add zero-lag and negate
    # the AR parameters should have the opposite sign
    ma = np.r_[1] # add zero-lag
    y = 0.05 + arma_generate_sample(ar, ma, scale = 0.1, nsample = 250)
    
    fig, ax = plt.subplots(nrows = 3, ncols = 1,figsize=(8,12))
    ax[0].plot(np.arange(1, 251, 1), y) # first subgraph
    ax[0].title.set_text(
        r'AR(1) Process with $\phi$ = {}'.format(phi)) # add the title
    
    theoretical_acf = phi ** np.arange(0, 16, 1)
    ax[1].scatter(np.arange(0, 16, 1), theoretical_acf)
    ax[1].set_ylim([-1.0, 1.0])
    ax[1].title.set_text('Theoretical ACF')
    
    plot_acf(y, lags = 15, ax = ax[2]) # second subgraph
    ax[2].title.set_text('Sample ACF')
    plt.show()
```

The process with $\phi = 0.5$ begins to demonstrate the autocorrelation we would hope for in an AR(1) process. Namely, there is moderate correlation with the first lag and approximately weak correlation with the second lag. The process itself still looks stationary.

## Problem 3

This exercise uses weekly data of Coca-Cola Co (KO) from the first
week of January, 2010 to the first week of March, 2021. Use the adjusted
closing prices to compute weekly cc returns. For all the sample auto-correlation functions (ACFs), set the maximum lag to 10.

### Part (a)

From the sample ACF, can we claim that the cc returns are white
noise? Conduct the Ljung-Box test (with lag = 10) and explain
your findings. (Hint: for the level-$\alpha$? Ljung-Box test, we reject the
null hypothesis if the p-value is less than $\alpha$?. Otherwise, we reject
the alternative hypothesis.)

#### Solution

```python
# Initial Imports
import yfinance as yf
import pandas as pd
from statsmodels.stats.diagnostic import acorr_ljungbox

df = yf.download("KO",
                start = "2010-01-01",
                end = "2021-03-06", 
                interval = '1wk')
df.reset_index(inplace = True) # convert index into a column
```

```python
df['Date'] = pd.to_datetime(df['Date']) # convert the strings to dates
cc = np.log(df['Adj Close']/df['Adj Close'].shift(1))
cc = np.array(cc)[1:]
```

```python
# Ljung Box Test
acorr_ljungbox(cc, lags=[10], return_df=True)
```

A Ljung-Box statistic of 23.988 implies that we would reject the null at the 1% significance level. This means that we reject the null hypothesis that our process is stationary. Therefore, we cannot claim that our cc returns are white noise since the autocorrelation is not zero (implies dependent time series process).

### Part (b)

Estimate an MA(1) model using the cc returns. From the sample
ACF of the fitted residuals, can we claim that MA(1) models the
(linear) dependence of the cc returns appropriately?

#### Solution

```python
from statsmodels.tsa.arima.model import ARIMA
res = ARIMA(cc, order=(0, 0, 1)).fit() # fit the MA(1) model
acorr_ljungbox(res.resid, lags=[10], return_df=True)
```

With a Ljung-Box statistic of 21.1688, we can reject the null (uncorrelated time series process) at the 5% level.  Since the autocorrelations are not 0, then we can claim that there is some linear dependence between cc returns.

### Part (c)

Find the optimal lag $\hat{q}_{\text{opt}}$ if we want to use MA(q) to model the cc returns. Then use teh data to estimate MA($\hat{q}_{\text{opt}}$). Do we reject the null hypothesis that $\theta_j = 0$ for each $j=1, \dots, \hat{q}_{\text{opt}}$?

#### Solution

```python
import pmdarima as pm

# MA Estimation
## initial guesses for p and q
MA_Est = pm.auto_arima(cc, start_p=0, start_q=1,
                      d=0, # non-seasonal difference order
                      max_p=0, max_q=10, # maximum p and q
                      stepwise=True)
print(MA_Est.summary())
```

It seems that the optimal number of lags for our model is MA(q) model is 1.

```python
# Model Coefficients
Est_html = MA_Est.summary().tables[1].as_html()
table = pd.read_html(Est_html, header=0, index_col=0)[0]
coef = table.iloc[:, 0].values
se = table.iloc[:, 1].values
p_vals = table.iloc[:, 3].values
print(coef)
print(se)
```

The above values are the coefficient and standard error values for the MA(1) model. Namely, we see a negative coefficient for the lag.

```python
print((coef - 0)/se)
print(p_vals) # more precise
```

From the above printout and table, we see that the intercept is not statistically significant. However, all other terms seem to be statistically significant. Therefore, we can reject the null that $\theta_1 = 0$.

### Part (d)

Find the sample ACF of the fitted residuals from part (c). From
the sample ACF of the fitted residuals, can we claim that MA($\hat{q}_{\text{opt}}$)
models the (linear) dependence of the cc returns appropriately?

#### Solution

```python
# Plotting
plt.figure()
plot_acf(MA_Est.resid(), lags = 10)
plt.show()
#Ljung-Box
print(acorr_ljungbox(MA_Est.resid(), lags=[10], return_df=True))
```

Based on the Ljung-Box statistic, we can reject the null hypothesis (stationarity) at the 5% level. Therefore, yes, we can claim that there is linear dependence.
